## 3. –ü–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ –º–æ–¥–µ–ª–∏ –±–∏–Ω–∞—Ä–Ω–æ–π –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏–∏ –∫–ª–∏–µ–Ω—Ç–æ–≤, —Ü–µ–ª–µ–≤–æ–π –ø—Ä–∏–∑–Ω–∞–∫ - —Ñ–∞–∫—Ç –æ—Ç—Ç–æ–∫–∞ –≤ —Å–ª–µ–¥—É—é—â–µ–º –º–µ—Å—è—Ü–µ –∏ –æ—Ü–µ–Ω–∫–∞ –º–µ—Ç—Ä–∏–∫ accuracy, precision –∏ recall.

````python
# —Ä–∞–∑–¥–µ–ª–∏–º –¥–∞–Ω–Ω—ã–µ –Ω–∞ –ø—Ä–∏–∑–Ω–∞–∫–∏ –∏ —Ü–µ–ª–µ–≤—É—é –ø–µ—Ä–µ–º–µ–Ω–Ω—É—é
X = df.drop(columns = ['churn','month_to_end_contract'])
y = df['churn']

# –¥–µ–ª–∏–º –≤—ã–±–æ—Ä–∫—É –Ω–∞ —Ç–µ—Å—Ç–æ–≤—É—é –∏ –æ–±—É—á–∞—é—â—É—é
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2, random_state=0, stratify = y)

# —Å—Ç–∞–Ω–¥–∞—Ä—Ç–∏–∑–∏—Ä—É–µ–º
Sc = StandardScaler()
X_sc = Sc.fit_transform(X)
X_train_sc = Sc.fit_transform(X_train)
X_test_sc = Sc.transform(X_test)
````

````python
# –†–∞—Å—á—ë—Ç LogisticRegression
model = LogisticRegression(solver='liblinear',random_state = 0)
model.fit(X_train_sc, y_train)
y_pred_lr = model.predict(X_test_sc)
y_probas_lr = model.predict_proba(X_test_sc)[:,1]

# –†–∞—Å—á—ë—Ç RandomForestClassifier
model = RandomForestClassifier(random_state = 0)
model.fit(X_train_sc, y_train)
y_pred_rfc = model.predict(X_test_sc)
y_probas_rfc = model.predict_proba(X_test_sc)[:,1]
````

````python
# –†–∞—Å—á—ë—Ç –º–µ—Ç—Ä–∏–∫ LogisticRegression
accuracy_lr = accuracy_score(y_test,y_pred_lr)
precision_lr = precision_score(y_test,y_pred_lr)
recall_lr = recall_score(y_test,y_pred_lr)
balanced_accuracy_lr = balanced_accuracy_score(y_test,y_pred_lr)
print('–ú–µ—Ç—Ä–∏–∫–∏ –ª–æ–≥–∏—Å—Ç–∏—á–µ—Å–∫–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏ —Å–æ —Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω—ã–º threshold')
print(f'accuracy: {accuracy_lr:.5f} \nprecision: {precision_lr:.5f}\nrecall: {recall_lr:.5f}\nbalanced_accuracy: {balanced_accuracy_lr:.5f}')
print()

# –†–∞—Å—á—ë—Ç –º–µ—Ç—Ä–∏–∫ RandomForestClassifier
accuracy_rfc = accuracy_score(y_test,y_pred_rfc)
precision_rfc = precision_score(y_test,y_pred_rfc)
recall_rfc = recall_score(y_test,y_pred_rfc)
balanced_accuracy_rfc = balanced_accuracy_score(y_test,y_pred_rfc)
print('–ú–µ—Ç—Ä–∏–∫–∏ RandomForestClassifier —Å–æ —Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω—ã–º threshold')
print(f'accuracy: {accuracy_rfc:.5f} \nprecision: {precision_rfc:.5f}\nrecall: {recall_rfc:.5f}\nbalanced_accuracy: {balanced_accuracy_rfc:.5f}')
````

- –û–±—Ä–∞—Ç–∏ –≤–Ω–∏–º–∞–Ω–∏–µ –Ω–∞ –∞—Ä–≥—É–º–µ–Ω—Ç stratify, –æ–Ω –ø–æ–∑–≤–æ–ª–∏—Ç —Å–æ—Ö—Ä–∞–Ω–∏—Ç—å –∏–∑–Ω–∞—á–∞–ª—å–Ω–æ–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ —Ç–∞—Ä–≥–µ—Ç–æ–≤ –≤–æ –≤—Å–µ—Ö –Ω–æ–≤—ã—Ö –¥–∞—Ç–∞—Å–µ—Ç–∞—Ö.  –°—É—â–µ—Å—Ç–≤—É—é—â–∏–π –¥–∏—Å–±–∞–ª–∞–Ω—Å –Ω–∏–∫—É–¥–∞ –Ω–µ –¥–µ–Ω–µ—Ç—Å—è, –Ω–æ –≤ –∫–∞–∂–¥–æ–º –¥–∞—Ç–∞—Å–µ—Ç–µ –æ–Ω –±—É–¥–µ—Ç –æ–¥–∏–Ω–∞–∫–æ–≤—ã–º. –í–æ–ø—Ä–æ—Å–∏–∫–∏. –ï—Å–ª–∏ –µ—Å—Ç—å –∂–µ–ª–∞–Ω–∏–µ –º–æ–∂–µ—à—å –æ—Ç–≤–µ—Ç–∏—Ç—å )


- —á–µ–º –ø–ª–æ—Ö–∞  –º–µ—Ç—Ä–∏–∫–∞ –∞ccuracy –¥–ª—è –¥–∞–Ω–Ω–æ–≥–æ –¥–∞—Ç–∞—Å–µ—Ç–∞? –∏ –Ω–∞ —á—Ç–æ —Å–¥–µ–ª–∞—Ç—å –∫–æ—Ä—Ä–µ–∫—Ü–∏—é —á—Ç–æ–±—ã –ø—Ä–∞–≤–∏–ª—å–Ω–æ –µ–µ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å


- –Ω–∞ —Ç–≤–æ–π –≤–∑–≥–ª—è–¥ –∫–∞–∫–∞—è –º–µ—Ç—Ä–∏–∫–∞  recall –∏–ª–∏ precision –∏ –æ—Ç–Ω–æ—Å–∏—Ç–µ–ª—å–Ω–æ –∫–∞–∫–æ–≥–æ –∫–ª–∞—Å—Å–∞ —è–≤–ª—è–µ—Ç—Å—è –±–æ–ª–µ–µ –≤–∞–∂–Ω–æ–π –≤ —ç—Ç–æ–º –ø—Ä–æ–µ–∫—Ç–µ, —Å —Ç–æ—á–∫–∏ –∑—Ä–µ–Ω–∏—è –±–∏–∑–Ω–µ—Å–∞?!


- –∫–∞–∫–∏–µ —Å–ø–æ—Å–æ–±—ã —Ç—ã –∑–Ω–∞–µ—à—å —á—Ç–æ–±—ã –ø–æ–¥–Ω—è—Ç—å  recall (precision)?


- –¥–æ–ø—É—Å—Ç–∏–º –Ω–∞–º –∏–∑–≤–µ—Å—Ç–Ω—ã —É–±—ã—Ç–∫–∏ –≤ —Å–ª—É—á–∞–∏ —É—Ö–æ–¥–∞ –∫–ª–∏–µ–Ω—Ç–∞ –∏ –∏–∑–≤–µ—Å—Ç–Ω—ã —Ä–∞—Å—Ö–æ–¥—ã —Å–≤—è–∑–∞–Ω–Ω—ã–µ —Å –ø–æ–ø—ã—Ç–∫–æ–π —É–¥–µ—Ä–∂–∞—Ç—å –∫–ª–∏–µ–Ω—Ç–∞ –æ—Ç —É—Ö–æ–¥–∞, –∏ –ø—Ä–∏ –Ω–∞—à–∏—Ö recall –∏ precision –æ–∫–∞–∑—ã–≤–∞–µ—Ç—Å—è —á—Ç–æ —Ä–∞—Å—Ö–æ–¥—ã –Ω–∞ —É–¥–µ—Ä–∂–∞–Ω–∏–µ –∫–ª–∏–µ–Ω—Ç–æ–≤ –≥–æ—Ä–∞–∑–¥–æ –Ω–∏–∂–µ —á–µ–º –ø—Ä–∏–±—ã–ª—å –æ—Ç —Ç–æ–≥–æ —á—Ç–æ –Ω–µ–∫–æ—Ç–æ—Ä—ã—Ö –∫–ª–∏–µ–Ω—Ç–æ–≤ —É–¥–∞–ª–æ—Å—å —É–¥–µ—Ä–∂–∞—Ç—å. –∫–∞–∫—É—é –º–µ—Ç—Ä–∏–∫—É —Å—Ç–æ–∏–ª–æ –ø–æ–ø—Ä–æ–±–æ–≤–∞—Ç—å –ø–æ–¥–Ω—è—Ç—å –≤ —ç—Ç–æ–º —Å–ª—É—á–∞–∏?

````python
# —Å–æ–∑–¥–∞–¥–∏–º —Å–ø–∏—Å–æ–∫ `threshold` –¥–ª—è –ø–æ–∏—Å–∫–∞ –æ–ø—Ç–∏–º–∞–ª—å–Ω–æ–≥–æ –∑–Ω–∞—á–µ–Ω–∏—è `recall`
thres_list = [x/100 for x in range(10, 50, 1)]
recall_lr_new = recall_lr
recall_rfc_new = recall_rfc
balanced_accuracy_lr_new = balanced_accuracy_lr
precision_lr_new = precision_lr
precision_rfc_new = precision_rfc
balanced_accuracy_lr_new = balanced_accuracy_rfc

threshold_lr_new = 0.5
threshold_rfc_new = 0.5

# –Ω–∞–π–¥—ë–º –ª—É—á—à–∏–π threshold –¥–ª—è –º–µ—Ç—Ä–∏–∫ c –¥–≤–æ–π–Ω—ã–º –≤–µ—Å–æ–º recall

for t_h in thres_list:
    y_pred_lr_exploring = [0 if i < t_h else 1 for i in y_probas_lr]
    y_pred_rfc_exploring = [0 if i < t_h else 1 for i in y_probas_rfc]
    
    recall_lr_exploring = recall_score(y_test,y_pred_lr_exploring)
    recall_rfc_exploring = recall_score(y_test,y_pred_rfc_exploring)
    
    precision_lr_exploring = precision_score(y_test,y_pred_lr_exploring)
    precision_rfc_exploring = precision_score(y_test,y_pred_rfc_exploring)
    
    balanced_accuracy_lr_exploring = balanced_accuracy_score(y_test,y_pred_lr_exploring)
    balanced_accuracy_rfc_exploring = balanced_accuracy_score(y_test,y_pred_rfc_exploring)
    
    if (2*recall_lr_exploring + precision_lr_exploring + balanced_accuracy_lr_exploring) > (2* recall_lr_new + precision_lr_new +balanced_accuracy_lr_exploring):
        recall_lr_new = recall_lr_exploring
        precision_lr_new = precision_lr_exploring
        balanced_accuracy_lr_new = balanced_accuracy_lr_exploring
        threshold_lr_new = t_h
    if (2*recall_rfc_exploring + precision_rfc_exploring + balanced_accuracy_rfc_exploring) > (2* recall_rfc_new + precision_rfc_new + balanced_accuracy_rfc_exploring):
        recall_rfc_new = recall_rfc_exploring
        precision_rfc_new = precision_rfc_exploring
        balanced_accuracy_rfc_new = balanced_accuracy_rfc_exploring
        threshold_rfc_new = t_h
        
print(f'–õ—É—á—à–µ–µ –∑–Ω–∞—á–µ–Ω–∏–µ threshold –¥–ª—è –õ–æ–≥–∏—Å—Ç–∏—á–µ—Å–∫–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏ = {threshold_lr_new}')
print(f'–∑–Ω–∞—á–µ–Ω–∏–µ balanced_accuracy = {balanced_accuracy_lr_new:.3f}')
print(f'–∑–Ω–∞—á–µ–Ω–∏–µ precision = {precision_lr_new:.3f}')
print(f'–∑–Ω–∞—á–µ–Ω–∏–µ recall = {recall_lr_new:.3f}')

print()
print(f'–õ—É—á—à–µ–µ –∑–Ω–∞—á–µ–Ω–∏–µ threshold –¥–ª—è Random Forest Classifier = {threshold_rfc_new}')
print(f'–∑–Ω–∞—á–µ–Ω–∏–µ balanced_accuracy = {balanced_accuracy_rfc_new:.3f}')
print(f'–∑–Ω–∞—á–µ–Ω–∏–µ precision = {precision_rfc_new:.3f}')
print(f'–∑–Ω–∞—á–µ–Ω–∏–µ recall = {recall_rfc_new:.3f}')
````

````python
print(f'–°—Ä–µ–¥–Ω–∏–π –æ—Ç—Ç–æ–∫ –ø–æ –¥–∞—Ç–∞—Å–µ—Ç—É: {df.churn.mean():.2%}')
````

–ü–æ –ø–æ–∫–∞–∑–∞—Ç–µ–ª—è–º ACCURACY, PRECISION –∏ RECALL –ª—É—á—à–µ –≤ –ø—Ä–æ–≥–Ω–æ–∑–µ –æ—Ç—Ç–æ–∫–∞ –∫–ª–∏–µ–Ω—Ç–æ–≤ –ø–æ–∫–∞–∑–∞–ª–∞ —Å–µ–±—è –õ–æ–≥–∏—Å—Ç–∏—á–µ—Å–∫–∞—è —Ä–µ–≥—Ä–µ—Å—Å–∏—è.

–û–¥–Ω–∞–∫–æ Accuracy –ø–æ–¥—Ö–æ–¥–∏—Ç –¥–ª—è –æ—Ü–µ–Ω–∫–∏ –∞–ª–≥–æ—Ä–∏—Ç–º–∞ –ø—Ä–æ–≥–Ω–æ–∑–∏—Ä–æ–≤–∞–Ω–∏—è –ø–ª–æ—Ö–æ, —Ç.–∫. –¥–æ–ª—è –æ—Ç—Ç–æ–∫–∞ –ø–æ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è–º —Ç–æ–ª—å–∫–æ 26,5%, –≤—ã–±–æ—Ä–∫–∞ –Ω–µ—Å–±–∞–ª–∞–Ω—Å–∏—Ä–æ–≤–∞–Ω–∞.

* –ú–µ—Ç—Ä–∏–∫—É –º–æ–∂–Ω–æ —Å–±–∞–ª–∞–Ω—Å–∏—Ä–æ–≤–∞—Ç—å –ø—Ä–æ—Å—á–∏—Ç–∞–≤ —Å—Ä–µ–¥–Ω–µ–µ –¥–ª—è 0 –∏ 1 `churn` –ø–æ –æ—Ç–¥–µ–ª—å–Ω–æ—Å—Ç–∏ –∏ —É—Å—Ä–µ–¥–Ω–∏–≤ —Ä–µ–∑—É–ª—å—Ç–∞—Ç
* –õ–∏–±–æ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å built-in –º–µ—Ç—Ä–∏–∫—É `balanced_accuracy()`

–í —Ä–∞–∑—Ä–µ–∑–µ —ç—Ç–æ–≥–æ –∏—Å—Å–ª–µ–¥–æ–≤–∞–Ω–∏—è –º–µ—Ç—Ä–∏–∫–∞ `recall` –º–Ω–µ –≤–∏–¥–∏—Ç—Å—è –±–æ–ª–µ–µ –≤–∞–∂–Ω–æ–π, —á–µ–º `precision`, —Ç.–∫. –ª—É—á—à–µ —Ä–∞—Å—à–∏—Ä–∏—Ç—å —Ü–µ–ª–µ–≤—É—é –∞—É–¥–∏—Ç–æ—Ä–∏—é –¥–ª—è —Å–Ω–∏–∂–µ–Ω–∏—è –æ—Ç—Ç–æ–∫–∞, —á–µ–º –ø–æ—Ç–µ—Ä—è—Ç—å —á–∞—Å—Ç—å –∫–ª–∏–µ–Ω—Ç–æ–≤ –æ—Ä–∏–µ–Ω—Ç–∏—Ä—É—è—Å—å —Ç–æ–ª—å–∫–æ –Ω–∞ –Ω–∞–¥—ë–∂–Ω–æ "–æ—Ç—Ç–µ–∫–∞—é—â–∏—Ö" –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π.

–ü—Ä–∏ —Å–Ω–∏–∂–µ–Ω–∏–∏ –ø–æ—Ä–æ–≥–∞ –æ—Ç–Ω–µ—Å–µ–Ω–∏—è –∫ –∫–ª–∞—Å—Å—É –¥–æ 0,27 –õ–æ–≥–∏—Å—Ç–∏—á–µ—Å–∫–∞—è —Ä–µ–≥—Ä–µ—Å—Å–∏—è –ø–æ–∑–≤–æ–ª—è–µ—Ç –æ—Ö–≤–∞—Ç–∏—Ç—å 94,3% –∫–ª–∏–µ–Ω—Ç–æ–≤ —Å —Ä–∏—Å–∫–æ–º –æ—Ç—Ç–æ–∫–∞.

–°–æ–≤–µ—Ç 1.

–¢–µ–º –Ω–µ –º–µ–Ω–µ–µ —Ö–æ—á—É –Ω–∞–ø–æ–º–Ω–∏—Ç—å –æ –ª–æ–≥–∏–∫–µ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏—è –¥–∞—Ç–∞—Å–µ—Ç–æ–≤ –∫–æ–≥–¥–∞ –ø—Ä–æ–±—É–µ–º –Ω–µ—Å–∫–æ–ª—å–∫–æ –º–æ–¥–µ–ª–µ–π –∏ –≤—ã–±–∏—Ä–∞–µ–º –ª—É—á—à—É—é:

### 4.2. –û–±—É—á–µ–Ω–∏–µ –º–æ–¥–µ–ª–∏ –∫–ª–∞—Å—Ç–µ—Ä–∏–∑–∞—Ü–∏–∏ –Ω–∞ –æ—Å–Ω–æ–≤–∞–Ω–∏–∏ –∞–ª–≥–æ—Ä–∏—Ç–º–∞ K-Means

````python
# –≤—ã–¥–µ–ª–∏–º 5 –∫–ª–∞—Å—Ç–µ—Ä–æ–≤ —Å –ø–æ–º–æ—â—å—é k_means –∏ –∑–∞–ø–∏—à–µ–º –∑–Ω–∞—á–µ–Ω–∏—è –∫–ª–∞—Å—Ç–µ—Ä–æ–≤ –≤ –∏—Å—Ö–æ–¥–Ω—ã–π –¥–∞—Ç–∞—Ñ—Ä–µ–π–º
k_model = KMeans(n_clusters=5, random_state=0)
clusters = k_model.fit_predict(X_sc)
df['km_clusters'] = clusters
print(df['km_clusters'].unique())
````

````python
# –ø–æ—Å—Ç—Ä–æ–∏–º —Ç–∞–±–ª–∏—Ü—É —Å–æ —Å—Ä–µ–¥–Ω–∏–º–∏ –∑–Ω–∞—á–µ–Ω–∏—è–º–∏ –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ –ø–æ –∫–ª–∞—Å—Ç–µ—Ä–∞–º
df_mean_by_clusters = pd.DataFrame()

for i in df['km_clusters'].unique():
    m = df[df['km_clusters'] == i].groupby(['km_clusters']).mean().T
    df_mean_by_clusters['–°—Ä–µ–¥–Ω–µ–µ –ø–æ –∫–ª.‚Ññ ' + str(i)] = m[i]
display(df_mean_by_clusters.style.background_gradient(cmap='coolwarm', axis=1).set_precision(2))
````
